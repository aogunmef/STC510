# Importing Datasets

I imported a data set that has all the total wages for a given zip code all across the US. In order to make the data sets match, I narrowed the dataframe for the wages to only include Phoenix as well as only zip codes that are not PO boxes

```python
crimes = pd.read_csv("crimestat.csv", delimiter=',', nrows=100000)
temp = pd.read_csv("temperature.csv", delimiter=",", low_memory=False)
temp = temp[["Date","Phoenix"]]
temp['Date']
```

```python
print(type(temp.Date[0]))
temp['Date'] = pd.to_datetime(temp['Date'])
print(type(temp.Date[0]))
temp.sort_values(by='Date')
temp
```

```python
crimes['Date'] = pd.to_datetime(crimes['Date'])
crimes
```

```python
fiteredTemp = temp.loc[(temp['Date'] >= '2016-01-01 00:00:00') & (temp['Date'] < '2016-12-31 00:00:00')]
fiteredTemp
```

```python
fiteredCrimes = crimes.loc[(crimes['Date'] >= '2016-01-01 00:00:00') & (crimes['Date'] < '2016-12-31 00:00:00')]
fiteredCrimes
```
